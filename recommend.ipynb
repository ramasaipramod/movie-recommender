```python
import numpy as np 
import pandas as pd
```

```python
credits = pd.read_csv('../Data/tmdb_5000_credits.csv')
movies = pd.read_csv('../Data/tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
# genres
# id
# keywords
# overview
# title
# cast 
# crew

movies = movies[['id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.reset_index(drop=True, inplace=True)
```

```python
movies.head()
```

```python
def rmvspace(L):
    newL = []
    for i in L:
        newL.append(i.replace(' ',''))
    return newL
```

```python
genres = [[eval(movies['genres'][i])[j]['name'] for j in range(len(eval(movies['genres'][i])))] for i in range(len(movies))]
keywords = [[eval(movies['keywords'][i])[j]['name'] for j in range(len(eval(movies['keywords'][i])))] for i in range(len(movies))]
```

```python
cast = []
for i in range(len(movies)):
    ith_cast = [' ']
    for j in range(3):
        try: 
            ith_cast[0] = eval(movies['cast'][i])[j]['name']
        except:
            None
    cast.append(ith_cast)
```

```python
director = []
for i in range(len(movies)):
    ith_director = [' ']
    for j in eval(movies['crew'][i]):
        if j['job']=='Director':
            ith_director[0] = j['name']
    director.append(ith_director)
```

```python
movies['genres'] = pd.Series(genres).apply(rmvspace)
movies['keywords'] = pd.Series(keywords).apply(rmvspace)
movies['cast'] = pd.Series(cast).apply(rmvspace)
movies['crew'] = pd.Series(director).apply(rmvspace)
```

```python
movies.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import *
```

```python
stemmer = PorterStemmer()

overview = []
for i in range(len(movies)):
    overview.append(movies['overview'][i].split())

ns_overview = [[word for word in overview[i] if word not in stopwords.words('english')] for i in range(len(movies))]

overview = []
for i in range(len(movies)):
    stemmed_words = []
    for j in range(len(ns_overview[i])):
        stemmed_words.append(stemmer.stem(ns_overview[i][j]))
    overview.append(stemmed_words)
```

```python
movies['overview'] = pd.Series(overview)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']

movies = movies.drop(columns=['overview', 'keywords', 'genres', 'cast', 'crew'])
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x))
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommender(movie_title):
    index = movies[movies['title'] == movie_title].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title) 
```

```python
recommender('Spectre')
```
